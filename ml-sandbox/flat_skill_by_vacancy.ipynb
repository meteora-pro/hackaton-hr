```python
import json
import os

VACANCIES_ITEMS_PATH = './vacancies/details'

files = os.listdir(VACANCIES_ITEMS_PATH)

domains = ['frontend', 'backend', 'product', 'data', 'designer', 'devops']

for domain in domains:
    domain_vacancies = {}

    for fileNum, fl in enumerate(files, start=1):
        if domain not in fl:
            continue
        print(f"load data for file {fileNum} of {len(files)} ", end='\r')
        # Открываем, читаем и закрываем файл
        f = open(f"{VACANCIES_ITEMS_PATH}/{fl}", encoding="utf8")
        print(f"read file ${f.name}")
        jsonText = f.read()
        f.close()
        chunkJson = json.loads(jsonText)
        if isinstance(chunkJson, dict):
            for num, key in enumerate(chunkJson.keys(), start=1):
                domain_vacancies[key] = chunkJson[key]
        else:
            print(f"file {fl} is not contain dict")

    domain_skills = {}

    for k, vacancy in domain_vacancies.items():
        if 'key_skills' not in vacancy:
            continue
        for skill in vacancy['key_skills']:
            if (skill['name'] in domain_skills):
                domain_skills[skill['name']] = domain_skills[skill['name']] + 1
            else:
                domain_skills[skill['name']] = 1

    uniq_skills = dict(reversed(sorted(domain_skills.items(), key=lambda item: item[1])))
    filtered = dict((k, v) for k, v in uniq_skills.items() if v > 1)
    fileName = f"./skills/{domain}_all.json"
    f = open(fileName, mode='w', encoding='utf8')
    f.write(json.dumps(filtered, ensure_ascii=False, indent=2))
    f.close()

    print(f"TOTAL VACATIONS: {len(domain_vacancies.keys())}")
```
